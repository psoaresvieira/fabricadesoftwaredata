```python
#Importando bibliotecas 

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#Criando variável e armazendando o dataframe nela. Além de usar a função read_csv para importar os dados.

df = pd.read_csv('c:\\Users\mvspb\dados_rh.csv')
df
```

```python
#Verificando o tamanho do dataframe (linhas, colunas), usando a função shape

df.shape

#Vemos que no dataframe há 1470 linhas e 35 colunas
```

```python
#Verificando se há dados nulos no dataframe, com a função null, e somando-os com a função sum

df.isnull().sum()

#Não há nenhum dado nulo neste dataframe 
```

```python
#Verificando quais tipos de dados estão presentes no dataframe, usando a função info

df.info()

#Aqui podemos ver especificamente quais são os tipos de dados de cada coluna do dataframe, os dados que estão presentes
#na tabela são: int64 e object. Sendo Int64 para receber inteiros e Object para receber strings
```

```python
#Agora vamos plotar/gerar os gráficos a partir de valores agrupados 
```

```python
#Tarefa de Plots 1 - Plotando um gráfico de barras que exiba a quantidade de funcionários em cada setor

#Agrupando os funcionários por setor. Aqui eu usei a função groupby para agrupar os funcionários por departamento 
#que estão sendo contados atráves da coluna contagem de funcionários, usando a função count e armazenando na variável agrupado1.

agrupado1 = df.groupby('Department')[['EmployeeCount']].count()

#Com a função groupby + a função size podemos ver os valores distintos da coluna department e sua contagem.

department_distinct = df.groupby('Department').size()
print(department_distinct)

#E finalmente iremos gerar o gráfico, vizualizando a quantidade de funcionários por cada setor. 
#Usando o plot para plotar o gráfico, o plt.legend para ajustar a legenda do gráfico, que estava muito dentro.
#E o plt.tilte para colocar um título no gráfico.

agrupado1.plot(kind = 'bar', color= 'tomato')
plt.legend(ncol=1, loc='upper right', bbox_to_anchor=(1.4, 1))
plt.title('Quantidade de Funcionários por Setor')

#Vemos que o departamento research & development é o com a maior quantidade de funcionários.
```

```python
#Tarefa de Plots 2 - Plotando um gráfico de linha que exiba a quantidade de funcionários para cada nível de escolaridade.

#Agrupando os Funcionário por nível de escolaridade. Aqui eu usei também a função groupby para agrupar os funcionários
#por nível escolar que estão sendo contados atráves da coluna contagem de funcionários, usando a função count. E armazenando na
#variável agrupado2.

agrupado2 = df.groupby('Education')[['EmployeeCount']].count()

#Com a função groupby + a função size podemos ver os valores distintos da coluna education e sua contagem.

education_distinct = df.groupby('Education').size()
print(education_distinct)

#E finalmente iremos gerar o gráfico, vizualizando a quantidade de funcionários por nível escolar. 
#Usando o plot para plotar o gráfico, o plt.legend para ajustar a legenda do gráfico, que estava muito dentro.
#E o plt.tilte para colocar um título no gráfico.

agrupado2.plot(color = 'green', marker = 'o')
plt.legend(ncol=1, loc='upper right', bbox_to_anchor=(1.4, 1))
plt.title('Quantidade de Funcionários por Nível de Escolaridade')

#Vemos que o nível 3 de escolaridade é o com a maior quantidade de funcionários.
```

```python
#Tarefa de Plots 3 - Plotando um gráfico de pizza que exiba a quantidade de funcionários que tem ou não tem atrito na empresa.

#Agrupando os Funcionário por atrito na empresa. Aqui eu usei novamente a função groupby para agrupar os funcionários
#por atrito na empresa que estão sendo contados atráves da coluna contagem de funcionários, usando a função count. E armazenando
#na variável agrupado3.

agrupado3 = df.groupby('Attrition')[['EmployeeCount']].count()

#Com a função groupby + a função size podemos ver os valores distintos da coluna atrittion e sua contagem.

atrittion_distinct = df.groupby('Attrition').size()
print(atrittion_distinct)

#E finalmente iremos gerar o gráfico, vizualizando a quantidade de funcionários por atrito na empresa. 
#Usando o plot para plotar o gráfico, o plt.legend para ajustar a legenda do gráfico, que estava muito dentro.
#E o plt.tilte para colocar um título no gráfico.

agrupado3.plot(kind = 'pie', subplots = True, colors = [
    'gray',
    'darkblue'
])
plt.legend(ncol=1, loc='upper right', bbox_to_anchor=(1.3, 1))
plt.title('Quantidade de Funcionários por Atrito na Empresa')

#Vemos que há mais funcionários sem nenhum atrito na empresa, do que funcionários com atrito.
```

```python
#Aqui eu plotei 3 gráfico no mesmo plano

#Usei o plt.figure que é uma função com a responsabilidade de ser 
#o contêiner de nível superior para todos os elementos do gráfico. 

#A função subplot adiciona subplot a uma figura atual na posição de grade especificada. Aqui iremos especificar a posição
#dos gráficos. Irá receber três inteiros separados que descrevem a posição da subtrama.

#E plotei cada gráfico usando o plot.nome do gráfico

plt.figure(1)
plt.subplot(2,2,1)
plt.barh(df['Attrition'], df['EmployeeCount'])
plt.subplot(2,2,2)
plt.plot(df['Education'], df['EmployeeCount'])
plt.subplot(2,2,3)
plt.barh(df['Department'], df['EmployeeCount'])
```

```python
#Esse é um histograma criado através da biblioteca seaborn e sua função histoplot.

srn.histplot(df.TotalWorkingYears, color="blue", kde=False).set(title='Total de Anos Trabalhados')

#Vendo-o podemos analisar que a faixa de 5 a 10 anos de trabalho é a parte que mais temos funcionários, ou seja,
#há mais funcionários com 5 a 10 anos de trabalho.
```
