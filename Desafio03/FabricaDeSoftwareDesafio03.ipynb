```python
#Importando as bibliotecas que serão usadas para realizar a task
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#Criando a váriavel df que armazenará a Importação dos dados.csv através da função read_csv do Pandas
df = pd.read_csv('c:\\Users\mvspb\dados.csv')
#Com a função head consigo ver as primeiras linhas do meu dataframe
df.head()
```

```python
#Com a função shape consigo ver o corpo do meu dataframe (suas linhas e colunas)
df.shape
#Podemos ver que existem 1000 linhas e 8 colunas
```

```python
#Antes de realizar as tarefas pedidas, irei trocar o nome da coluna race/ethnicity por apenas ethnicity, 
#usando a função rename, para renomear
df.rename(columns = {'race/ethnicity': 'ethnicity'}, inplace=True)
```

```python
#Agora iremos ver se há algum valor null no dataframe, usando a função isnull + a sum.
df.isnull().sum()
#Não há valores null nesse dataframe
```

```python
#Tarefa 1 
#Antes de qualquer agrupamento, calcule quais os valores mínimos, máximos e médios de pontuação de cada matéria.

#Para começar irei usar o describe para descrever cada matéria
df.describe()
#Vemos que os valores máx de cada uma é (MATH = 100, READING = 100 e WRITING = 100)
#-------------------------------------------------------------------------------------
#Vemos que os valores min de cada uma é (MATH = 0, READING = 17, WRITING = 10)
#-------------------------------------------------------------------------------------
#Vemos que as médias de cada uma é (MATH = 66.089, READING = 69.169, WRITING = 68.054)
```

```python
#Tarefa 2
#Agrupe o banco de dados a partir da rça/etnia dos alunos,
#e plote um gráfico com a nota média em Matemática para cada grupo disponível

#Começaremos criando a variável 'agrupado' e armazenaremos nela o agrupamento a partir da raça/etnia, usando a função groupby

agrupados = df.groupby('ethnicity')[['math score']].mean()

#Após o passo anterior, iremos plotar o gráfico de barras, que irá mostrar o agrupamento feito anteriormente, 
#com o plt.title damos um título para o nosso gráfico

agrupados.plot(kind = 'bar', color = 'orange')
plt.title('Nota Média Matemática de cada Etnia')

#A partir dessa análise, podemos ver que a etnia grupo E tem uma média maior em Matemática, em relação as outras
```

```python
#Tarefa 3
#Agrupe o banco de dados a partir do Nível de Escolaridade dos pais, e plote um gráfico com a nota média em Escrita para cada
#grupo disponível

#Começaremos criando a variável 'agrupado2' e armazenaremos nela o agrupamento a partir do Nível de Escolaridade dos pais
#, usando a função groupby

agrupado2 = df.groupby('parental level of education')[['writing score']].mean()

#Após o passo anterior, iremos plotar o gráfico de pizza, que irá mostrar o agrupamento feito anteriormente, 
#com o plt.title damos um título para o nosso gráfico
#A função legend foi algo que eu pesquisei para mover a legenda, pois estava na frente do gráfico.

agrupado2.plot(kind = 'pie', subplots=True)
plt.legend(ncol=1, loc='upper right', bbox_to_anchor=(2, 1))
plt.title('Nota Média de Escrita de cada casal de pais')
```

```python
#Ainda na Tarefa 3 
#Com o gráfico de pizza não foi possível analisar os dados, então eu irei usar o gráfico de barras na horizontal

agrupado2.plot(kind= 'barh', color='orange')
plt.legend(ncol=1, loc='upper right', bbox_to_anchor=(1.4, 1))
plt.title('Nota Média de Escrita de cada casal de pais')

#A partir dessa análise, podemos ver que o nível 'master's tem uma média maior em Escrita, em relação aos outros
```

```python
#Tarefa 4
#Agrupe o banco de dados a partir da coluna referente à participação em um curso preparatório para os testes, 
#e plote um gráfico com a nota média em Leitura para cada grupo disponível.

#Começaremos criando a variável 'agrupado3' e armazenaremos nela o agrupamento a partir do curso preparatório para os testes
#, usando a função groupby

agrupado3 = df.groupby('test preparation course')[['reading score']].mean()

#Após o passo anterior, iremos plotar o gráfico de barras, que irá mostrar o agrupamento feito anteriormente, 
#com o plt.title damos um título para o nosso gráfico

agrupado3.plot(kind = 'bar')
plt.title('Nota Média Leitura de quem realizou os cursos preparatórios para os testes e quem não realizou')

#A partir dessa análise, podemos ver que quem fez o curso preparatório tem uma média maior em Leitura, em relação a quem não fez
```
